```python
#creating a new columns
```

```python
fb['Price1'] = fb['Close'].shift(-1)
fb
```

```python
# price diiference with columns price diff( close price of tomorrow- close price of today)
```

```python
fb['PriceDiff']= fb['Price1']-fb['Close']
fb.head()
```

```python
# create daily return price diff/ close today  
```

```python
fb['Return'] =fb['PriceDiff']/fb['Close']
fb.head()
```

```python
# create direction rule  Pricediff>0 Up 1 , Price diff,<=0  down -1
```

```python
fb['Direction']=[1 if fb.loc[ei, 'PriceDiff']>0
                 else-1 for ei in fb.index]
fb.head()
```

```python
fb['Average3'] =(fb['Close']+ fb['Close'].shift(1)+fb['Close'].shift(2))/3
fb.head()
```

```python
# calculate moving average using.rolling()
```

```python
fb['MA40']=fb ['Close'].rolling(40).mean()
fb['MA200']= fb['Close'].rolling(200).mean()
```

```python
fb['Close'].plot()
fb['MA40'].plot()
```
